```python
import pandas as pd
pd.options.display.max_columns = None
pd.options.display.expand_frame_repr = False

df=pd.read_excel(r'gti.xlsx', sheet_name='Gay Travel Index')

print(df.head())
print(df.info())
```

```python
column_names=list(df.columns.values)

#loop through each column and print count of nulls and zeroes
for item in column_names:
    print(item, 'nulls:', df[item].isna().sum())
    print(item, 'dashes:', df[item].isin(['-']).sum())
```

```python
#make new df1 to preserve original df
df1=df

#replace all dashes and nulls with 0
for i in column_names[4:]:
    df1.loc[df1[i] == '-', i] = 0
    df1.loc[df1[i].isna(), i] = 0

#check again
for item in column_names:
    print(item, 'nulls:', df1[item].isna().sum())
    print(item, 'dashes:', df1[item].isin(['-']).sum())
```

```python
print(df1.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

numdf=df1[column_names[4:]]
print(numdf)

plt.figure(figsize=(20, 5))
plt.xticks(rotation=45)
sns.boxplot(data=numdf)
plt.ylim(-6, 6)
plt.show()
```
